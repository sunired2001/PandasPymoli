```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "c:/classwork/pandas/purchase_data.csv"

players_demogr=pd.DataFrame({})
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_datacp=purchase_data.copy()


```

## Player Count

* Display the total number of players


```python
purchase_datadrop=pd.DataFrame({})

purchase_datadrop=purchase_datacp.drop_duplicates(["SN","Age","Gender"])
#df.index is index column
total_rowsgender=len(purchase_datadrop.index)

purchase_data["Total Players"]=total_rowsgender

purchase_players=pd.DataFrame({"Total Players":purchase_data["Total Players"]})
del purchase_data["Total Players"]
purchase_players.head(1)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data["Numbr Of Unique Items"]=len(purchase_data["Item ID"].unique().tolist())

players_demogr=pd.DataFrame({"Numbr Of Unique Items":purchase_data["Numbr Of Unique Items"]})

totalrows=len(purchase_data.index)
sumprice=purchase_data["Price"].sum()
players_demogr["Average Price"]=sumprice/totalrows
players_demogr["Number Of Purchases"]=totalrows
players_demogr["Total Revenue"]=sumprice
 
players_demogr["Total Revenue"]=players_demogr["Total Revenue"].map("${:,}".format)
players_demogr["Average Price"]=players_demogr["Average Price"].map("${:.2f}".format)
players_demogr.head(1)

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python



purchaseddatagrouped=purchase_datadrop.groupby(["Gender"]).agg({"Gender":['count',lambda x: ((x.count()/total_rowsgender)*100)]})

purchaseddatagrouped.columns = ["_".join(x) for x in purchaseddatagrouped.columns.ravel()]

purchaserename=purchaseddatagrouped.rename(columns={"Gender_count":"Total Count","Gender_<lambda>":"Percentage Of Players"})

purchaserename["Percentage Of Players"]=purchaserename["Percentage Of Players"].map("{:.2f}".format)
purchaserename.sort_values(by="Total Count",ascending=False,inplace=True)
purchaserenamecp=purchaserename.copy()
del purchaserenamecp.index.name
purchaserenamecp.head(20)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchaserename.head(20)
```

```python
purchase_dataindex=purchase_datacp.set_index("Gender")
#merge to get total count of gender from purchaserename and product info from purchase_dataindex
purcahsemerge=pd.merge(purchase_dataindex,purchaserename,on="Gender")

del purcahsemerge["SN"]
del purcahsemerge["Age"]
del purcahsemerge["Item ID"]
del purcahsemerge["Item Name"]
del purcahsemerge["Percentage Of Players"]
purchasereset=purcahsemerge.reset_index()
def avgPrice(x):
    # x is a DataFrame of group values
    x['Avg Total Purchase per Person'] = x['Price'].sum()/x['Total Count']
    return x
puravgprice=purchasereset.groupby("Gender").apply(avgPrice)
purdropdupli=puravgprice.drop_duplicates(["Gender","Total Count","Avg Total Purchase per Person"])
purchaseddatagrouped=purchasereset.groupby("Gender").agg({"Purchase ID":'count',"Price":['sum','mean']})


purchaseddatagrouped.columns=purchaseddatagrouped.columns.map('_'.join)

purchaserename=purchaseddatagrouped.rename(columns={"Purchase ID_count":"Purchase Count","Price_mean":"Average Purchase Price","Price_sum":"Total Purchase Value"})

purchaserename["Average Purchase Price"]=purchaserename["Average Purchase Price"].map("${:.2f}".format)
purchaserename["Total Purchase Value"]=purchaserename["Total Purchase Value"].map("${:,.2f}".format)
purcahsemerge1=pd.merge(purchaserename[["Purchase Count","Average Purchase Price","Total Purchase Value"]],purdropdupli[["Gender","Avg Total Purchase per Person"]],on="Gender")

purana=purcahsemerge1.set_index("Gender")
purana["Avg Total Purchase per Person"]=purana["Avg Total Purchase per Person"].map("${:,.2f}".format)
purana.sort_index(inplace=True)
purana
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

bins=[0,9,14,19,24,29,34,39,200]
binlabels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_datadrop["Age Limit"]=pd.cut(purchase_datadrop["Age"],bins,labels=binlabels)

purchase_datadrop.head(10)
puragedemo=purchase_datadrop.groupby("Age Limit").agg({"Age":['count',lambda x: ((x.count()/total_rowsgender)*100)]})

puragedemo.columns=puragedemo.columns.map('_'.join)

#puragedemo.to_csv("c:/classwork/purc.csv")
purageren=puragedemo.rename(columns={"Age_count":"Total Count","Age_<lambda>":"Percentage of Players"})
purageren["Percentage of Players"]=purageren["Percentage of Players"].map("{:,.2f}".format)
puragerendropin=purageren.copy()
del puragerendropin.index.name
puragerendropin
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

purchase_data["Age Limit"]=pd.cut(purchase_data["Age"],bins,labels=binlabels)
purmerge=pd.merge(purageren,purchase_data,on="Age Limit")

purchasereset=purmerge.reset_index(drop=True)
def avgPrice(x):
    # x is a DataFrame of group values
    x['Avg Total Purchase per Person'] = x['Price'].sum()/x['Total Count']
    return x
puravgprice=purchasereset.groupby("Age Limit").apply(avgPrice)
purdropdupli=puravgprice.drop_duplicates(["Age Limit","Total Count","Avg Total Purchase per Person"])
del purdropdupli["Total Count"]
del purdropdupli["Percentage of Players"]
del purdropdupli["Purchase ID"]
del purdropdupli["SN"]
del purdropdupli["Age"]
del purdropdupli["Gender"]
del purdropdupli["Item ID"]
del purdropdupli["Item Name"]
del purdropdupli["Price"]
del purdropdupli["Numbr Of Unique Items"]
#del purdropdupli["index"]
purdropdupli.to_csv("c:/classwork/purchase_datadrop.csv")
purdropdupli["Avg Total Purchase per Person"]=purdropdupli["Avg Total Purchase per Person"].map("${:,.2f}".format)

purchaseddatagrouped=purmerge.groupby("Age Limit").agg({"Purchase ID":'count',"Price":['mean','sum']})
#purchaseddatagrouped=purmerge.groupby("Age Limit").agg({"Purchase ID":{"Purchase Count":'count'},"Price":{"Average Purchase Price":'mean',"Total Purchase Value":'sum'}})
purchaseddatagrouped
purchaseddatagrouped.columns = ["_".join(x) for x in purchaseddatagrouped.columns.ravel()]
purchaserename=purchaseddatagrouped.rename(columns={"Purchase ID_count":"Purchase Count","Price_mean":"Average Purchase Price","Price_sum":"Total Purchase Value"})

purchaserename["Average Purchase Price"]=purchaserename["Average Purchase Price"].map("${:.2f}".format)
purchaserename["Total Purchase Value"]=purchaserename["Total Purchase Value"].map("${:,.2f}".format)
purmerge=pd.merge(purchaserename,purdropdupli,on="Age Limit")

purmerge.set_index("Age Limit",inplace=True)
del purmerge.index.name
#purmerge
purchaserename
```
